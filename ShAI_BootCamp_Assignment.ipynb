# **Task \#3**

```python
'''
BasePay                609
OvertimePay              4
OtherPay                 4
Benefits             36163
Since the data is skewed (will see that in figures in task 4) I've decided to fill the missing data with the median
so I won't affect the distrubution of the data
'''
df['BasePay'].fillna(np.median(df['BasePay']), inplace = True)
df['OvertimePay'].fillna(np.median(df['OvertimePay']), inplace = True)
df['OtherPay'].fillna(np.median(df['OtherPay']), inplace = True)
df['Benefits'].fillna(np.median(df['Benefits']), inplace = True)
```

```python
'''
Notes               148654
Status              148654

These columns are empty so its better to drop them (or you can give them to the client to fill them)
'''
df.drop({"Notes", "Status"}, axis = 1, inplace = True)
df.head()
```

# **Task #4**

```python
import matplotlib.pyplot as plt

lst = ['BasePay', 'OvertimePay', 'OtherPay', 'Benefits', 'TotalPay', 'TotalPayBenefits']

for i in lst:
  plt.figure(figsize=(8, 6))
  plt.hist(df[i], bins = 20)
  plt.title(f'Histogram for {i}')
  plt.xlabel(i)
  plt.ylabel('Frequency')
  plt.show()
```

```python
'''
The ouput pie chart wasn't clear since there are 2159 Job Title
we can use a thresold to the jobtitle, any jobtitle that does not satisfy the condition will be changed to other
'''
x = df['JobTitle'].value_counts()

plt.figure(figsize=(10, 6))
plt.pie(x, labels = x.index)
plt.show()
```

```python
import seaborn as sns
correlation_matrix = df.corr()

# Plotting a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=.5)
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['TotalPayBenefits'], df['TotalPay'])
plt.title('Scatter Plot: TotalPay vs. TotalPayBenefits')
plt.xlabel('TotalPayBenefits')
plt.ylabel('TotalPay')
plt.show()
```
